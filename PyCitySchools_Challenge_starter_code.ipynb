```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to change the path if needed.)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read the School Data and Student Data and store into a Pandas DataFrame
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Cleaning Student Names and Replacing Substrings in a Python String
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# Iterate through the words in the "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")

# Check names.
student_data_df.head(10)
```

## Deliverable 1: Replace the reading and math scores.

### Replace the 9th grade reading and math scores at Thomas High School with NaN.

```python
# Install numpy using conda install numpy or pip install numpy. 
# Step 1. Import numpy as np.
import numpy as np
```

```python
# Step 2. Use the loc method on the student_data_df to select all the reading scores from the 9th grade at Thomas High School and replace them with NaN.
student_data_df.loc[(student_data_df["school_name"] == "Thomas High School") & (student_data_df["grade"] == "9th"),["reading_score"]] = "NaN"

student_data_df["reading_score"] = pd.Series(student_data_df["reading_score"], dtype=np.dtype("float64"))
```

```python
# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

## Math and Reading Scores by Grade

```python
# Create a Series of scores by grade levels using conditionals.
ninth_grade_students = student_data_df.loc[student_data_df["grade"] == "9th"]
tenth_grade_students = student_data_df.loc[student_data_df["grade"] == "10th"]
eleventh_grade_students = student_data_df.loc[student_data_df["grade"] == "11th"]
twelfth_grade_students = student_data_df.loc[student_data_df["grade"] == "12th"]

# Group each school Series by the school name for the average math score.
ninth_grade_student_math_scores = ninth_grade_students.groupby("school_name")["math_score"].mean()
tenth_grade_student_math_scores = tenth_grade_students.groupby("school_name")["math_score"].mean()
eleventh_grade_student_math_scores = eleventh_grade_students.groupby("school_name")["math_score"].mean()
twelfth_grade_student_math_scores = twelfth_grade_students.groupby("school_name")["math_score"].mean()

# Group each school Series by the school name for the average reading score.
ninth_grade_student_reading_scores = ninth_grade_students.groupby("school_name")["reading_score"].mean()
tenth_grade_student_reading_scores = tenth_grade_students.groupby("school_name")["reading_score"].mean()
eleventh_grade_student_reading_scores = eleventh_grade_students.groupby("school_name")["reading_score"].mean()
twelfth_grade_student_reading_scores = twelfth_grade_students.groupby("school_name")["reading_score"].mean()
```

```python
# Combine each Series for average math scores by school into single data frame.

school_by_math_score_df = pd.concat([ninth_grade_student_math_scores,tenth_grade_student_math_scores ,eleventh_grade_student_math_scores ,twelfth_grade_student_math_scores], axis =1 )
school_by_math_score_df = school_by_math_score_df.set_axis(["Ninth(M)","Tenth(M)","Eleventh(M)","Twelfth(M)"],axis = 1)
```

```python
# Combine each Series for average reading scores by school into single data frame.

school_by_reading_score_df = pd.concat([ninth_grade_student_reading_scores,tenth_grade_student_reading_scores ,eleventh_grade_student_reading_scores ,twelfth_grade_student_reading_scores], axis =1 )
school_by_reading_score_df = school_by_reading_score_df.set_axis(["Ninth(R)","Tenth(R)","Eleventh(R)","Twelfth(R)"],axis = 1)
```

```python
# Format each grade column.
school_by_math_score_df["Ninth(M)"] = school_by_math_score_df["Ninth(M)"].map("{:.2f}".format)
school_by_math_score_df["Tenth(M)"] = school_by_math_score_df["Tenth(M)"].map("{:.2f}".format)
school_by_math_score_df["Eleventh(M)"] = school_by_math_score_df["Eleventh(M)"].map("{:.2f}".format)
school_by_math_score_df["Twelfth(M)"] = school_by_math_score_df["Twelfth(M)"].map("{:.2f}".format)

school_by_math_score_df["Ninth(M)"] = school_by_math_score_df["Ninth(M)"].astype(float)
school_by_math_score_df["Tenth(M)"] = pd.to_numeric(school_by_math_score_df["Tenth(M)"])
school_by_math_score_df["Eleventh(M)"] = pd.to_numeric(school_by_math_score_df["Eleventh(M)"])
school_by_math_score_df["Twelfth(M)"] = pd.to_numeric(school_by_math_score_df["Twelfth(M)"])
# Format each grade column.
school_by_reading_score_df["Ninth(R)"] = school_by_reading_score_df["Ninth(R)"].map("{:.2f}".format)
school_by_reading_score_df["Tenth(R)"] = school_by_reading_score_df["Tenth(R)"].map("{:.2f}".format)
school_by_reading_score_df["Eleventh(R)"] = school_by_reading_score_df["Eleventh(R)"].map("{:.2f}".format)
school_by_reading_score_df["Twelfth(R)"] = school_by_reading_score_df["Twelfth(R)"].map("{:.2f}".format)

school_by_reading_score_df["Ninth(R)"] = school_by_reading_score_df["Ninth(R)"].astype(float)
school_by_reading_score_df["Tenth(R)"] = pd.to_numeric(school_by_reading_score_df["Tenth(R)"])
school_by_reading_score_df["Eleventh(R)"] = pd.to_numeric(school_by_reading_score_df["Eleventh(R)"])
school_by_reading_score_df["Twelfth(R)"] = pd.to_numeric(school_by_reading_score_df["Twelfth(R)"])

print(school_by_math_score_df.dtypes)
```

```python
# Remove the index.
school_by_math_score_df.reset_index(drop=True, inplace=True) 
school_by_reading_score_df.reset_index(drop=True, inplace=True) 


# Display the data frame
print(school_by_math_score_df)
print(school_by_reading_score_df)
```
